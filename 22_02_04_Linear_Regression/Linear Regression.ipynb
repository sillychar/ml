```python
##### >>Gradient Descent:
    -Step which is drawn with the help of partial derivative by 
    optimizing error >> Learning step
    -Learning step will become smaller and smaller in every step

>>Errors:
    Residual error (SSE/RSS)= actual - pred
    Regression error (SSR)= pred - mean
    total error (SST)= Residual + Regression
                =   actual - mean

>>Coeff of Determination R^2:
    -To evaluate the model/ to check the reliability of the model
    
    -If we have 1 ind var:
                    -Coefficient of Determination
                    - R * R
    -If we have more than 1 ind var :
                    Coefficient of Multiple Determination
    
    -How much of total variance is there in Y has been explained by our model  
    
    -measure is coeff. of determination
    
    -R-square(R^2) ranges between 0 and 1
    
    -evaluate the performance of the model
    
    -it will define how our model is

    -lesser the variance/residual to explain the model, better the model
    
    -R^2= 1- SSE/SST
        = 1- (sum(act - pred)^2)/(sum(act - mean)^2)
        
    -Highest value of R^2 can be 1, accuracy will be 100%, 
    
    -when SSE = 0
        R^2 = 1
        
    -When SST > SSE 
        R^2 = 0 to 1
        
    -When SSE > SST 
        R^2 = -ve
        
    -R^2 is used to check goodness of model
    
    ex. :
        if we have 10 ind var= R^2 = 0.75
           if we add 11th var as good pred = R =0.85
              R^2 = 0.78 
        
           if we add 11th var as bad pred = R = 0.3
              R^2 = 0.755
        
    -More the data , more accuracy
    -more variable, more accuarcy
        
    -R^2 is not considered as very good metric for evaluating model
     because it gets impacted by bad pred
        
    -So we make use of Adjusted R^2  
        
        
        
    
    
Adjusted R^2( A-R^2)

```
