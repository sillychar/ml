```python
import pandas as pd
import numpy as np
from scipy import stats
```

## Trimming(remove data)

```python
array1 = np.array([20,21,24,25,22,20,200,210,25,30,26,28])
array1
```

```python
array2 = np.random.randint(20,30, size= 12)
array2
```

```python
z_score = np.abs(stats.zscore(array1))
print( z_score)
threshold = 2
out_index = np.where(z_score > threshold)
outliers_list=array1[out_index]
print('Outliers in dataset:', outliers_list)
```

```python
z_score = np.abs(stats.zscore(array2))
print( z_score)
threshold = 2
out_index = np.where(z_score > threshold)
outliers_list=array2[out_index]
print('Outliers in dataset:', outliers_list)
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.describe()
```

```python
q1 = df['Age'].quantile(0.25)
q2 = df['Age'].quantile(0.5)
q3 = df['Age'].quantile(0.75)
print(f'q1 == {q1}, q2== {q2},q3 =={q3}')

iqr = q3 -q1
print('IQR is :',iqr)
```

```python
upper_limit = q3 + (1.5 * iqr)
lower_limit = q1 - (1.5 * iqr)

print('Upper limit is: ', upper_limit)
print('Lower limit is: ', lower_limit)
```

```python
# To remove extreme outliers only


# upper_limit = q3 + (3 * iqr)
# lower_limit = q1 - (3 * iqr)

# print('Upper limit is: ',upper_limit)
# print('Lower limit is: ', lower_limit)
```

```python
df[(df['Age']>upper_limit) | (df['Age'] < 10)]
```

```python
outliers_indices = df[(df['Age']>upper_limit) | (df['Age'] < lower_limit)].index
outliers_indices
```

```python
df1 = df.drop(outliers_indices, axis = 0)
df1
```
